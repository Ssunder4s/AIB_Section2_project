```python
import os
import pandas as pd
from glob import glob
from tqdm import tqdm
from tqdm.contrib import tzip

WD = 'C:\\Users\Sunwoo\Desktop\업무\code\[코드스테이츠] AIB16\AI_16_허선우_Section2_Project\Data'
os.chdir(WD)
```

```python
region = pd.Series(sorted(glob("데이터/*.xlsx"))).str.split('_').str[-1].str.split('.').str[0].to_list()

def make_dataset(wd, region):
    
    wd_list = sorted(glob(wd))
    df = pd.DataFrame()
    
    for fn, re in zip(wd_list, region):
        
        df_new = pd.read_excel(fn, index_col = 0)
        df_new['region'] = re
        df = pd.concat([df, df_new])
        
    return df

dataset_info = make_dataset(wd = "데이터/*.xlsx", region=region)
dataset_info.to_csv('dataset_info.csv')

pd.read_csv('dataset_info.csv')
```

```python
dataset_info = pd.read_csv('dataset_info.csv')
infodata = dataset_info.query('사업자등록번호 == 사업자등록번호' and '업소상태 == 업소상태')[['사업자등록번호', '법인여부', 'region', '업소상태',  '판매방식', '취급품목']]
infodata.drop_duplicates(inplace = True)
infodata.reset_index(inplace = True, drop = True)
```

```python
#취급 품목 추가
item_1 = pd.get_dummies(infodata['취급품목'].str.split('/ ').str[0])
item_2 = pd.get_dummies(infodata['취급품목'].str.split('/ ').str[1])

item_table = (item_1 + item_2).drop(columns = ['null', '기타'])
item_table['기타'] = (item_table.sum(axis=1) == 0).astype('int')

infodata = pd.concat([infodata, item_table], axis = 1)
infodata = infodata.iloc[infodata['사업자등록번호'].drop_duplicates().index].reset_index(drop=True)
infodata.drop(columns =['취급품목'], inplace = True)

# 판매방식
channel_1 = pd.get_dummies(infodata['판매방식'].str.split('/ ').str[0])
channel_2 = pd.get_dummies(infodata['판매방식'].str.split('/ ').str[1])

channel_table = (channel_1 + channel_2)

infodata = pd.concat([infodata, channel_table], axis = 1)
infodata.drop(columns =['판매방식', '법인여부'], inplace = True)

# 업소상태
rep_dict = {
    '통신판매업 신고': 1,
    '영업재개':1, 
    '통신판매업 폐업':0, 
    '통신판매업 휴업':0, 
    '직권말소' : 0, 
    '직권취소' : 0,
    '타시군구전입' : 1}
infodata['업소상태'] = infodata['업소상태'].replace(rep_dict)



# 수도권 여부
infodata['수도권여부'] = \
                    ((infodata['region'] == '서울특별시').astype('int') + \
                    (infodata['region'] == '경기도').astype('int') + \
                    (infodata['region'] == '인천광역시').astype('int'))
infodata.drop(columns =['region'], inplace = True)
```
